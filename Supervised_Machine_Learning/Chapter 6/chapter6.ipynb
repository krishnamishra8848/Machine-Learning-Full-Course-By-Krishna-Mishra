```python
import pandas as pd
import numpy as np

df=pd.read_csv('laptop_data.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
# Handling Categorical & Numerical Data Issues

# In our Dataset there are lot of issus
# ScreenResolution should be int but it is object
# Ram should be int but it is object
# Memory should be int but it is object
# Weight should be in float but it is object
# Price should be int but it is float
# Unnamed: 0 is unwanted column we should remove it
# Mixed Data in ScreenResolution,cpu
```

```python
# Fixing all the issues

# RAM should be int so we remove GB From it and convert it to Int
# Remove 'GB' and convert the column to integers
df['Ram'] = df['Ram'].str.replace('GB', '').astype(int)
```

```python
df.head()
```

```python
# Remove 'kg' (with or without space) from the 'Weight' column and convert to float
df['Weight'] = df['Weight'].str.replace(r'\s?kg', '', regex=True).astype(float)
```

```python
df.head()
```

```python
# Convert 'Price' column to integers
df['Price'] = df['Price'].astype(int)
```

```python
df.head()
```

```python
# Remove the 'Unnamed: 0' column
df = df.drop(columns=['Unnamed: 0'])
```

```python
df.head()
```

```python
# Remove non-numeric characters (like 'GB') from the 'Memory' column
df['Memory'] = df['Memory'].str.replace(r'\D', '', regex=True).astype(int)
```

```python
df.head()
```

```python
df['Company'].value_counts()
```

```python
# In Company We can see that some laptop are so many in our data while some are very low
# In this condition we make less value as name Other

# Count occurrences of each company
company_counts = df['Company'].value_counts()

# Replace companies with less than 10 occurrences with 'Other'
df['Company'] = df['Company'].apply(lambda x: x if company_counts[x] >= 10 else 'Other')
```

```python
df['Company'].value_counts()
```

```python
df['TypeName'].value_counts()
```

```python
# Count occurrences of each type
type_counts = df['TypeName'].value_counts()

# Replace types with fewer than 50 occurrences with 'Other'
df['TypeName'] = df['TypeName'].apply(lambda x: x if type_counts[x] >= 50 else 'Other')


```

```python
df['TypeName'].value_counts()
```

```python
df['OpSys'].value_counts()
```

```python
# Create a mapping for the OS categories
os_mapping = {
    'Windows 10': 'Windows',
    'Windows 7': 'Windows',
    'Windows 10 S': 'Windows',
    'Mac OS X': 'Mac',
    'macOS': 'Mac',
    'Linux': 'Linux'
}

# Apply the mapping to the 'OpSys' column, replacing other OS with 'Other'
df['OpSys'] = df['OpSys'].apply(lambda x: os_mapping.get(x, 'Other'))


```

```python
df['OpSys'].value_counts()
```

```python
df['Gpu'].value_counts()
```

```python
df['Gpu'].unique()
```

```python
# Function to extract the brand name from the GPU
def extract_gpu_brand(gpu):
    if 'Nvidia' in gpu:
        return 'Nvidia'
    elif 'AMD' in gpu:
        return 'AMD'
    elif 'Intel' in gpu:
        return 'Intel'
    else:
        return 'Other'

# Apply the function to the 'Gpu' column
df['Gpu'] = df['Gpu'].apply(extract_gpu_brand)
```

```python
df['Gpu'].value_counts()
```

```python
# Remove Other bcz its only 1 in data
# Remove rows where 'Gpu' is 'Other'
df = df[df['Gpu'] != 'Other']
```

```python
df['Gpu'].value_counts()
```

```python
df.head()
```

```python
df['Cpu'].value_counts()
```

```python
df['Cpu'].unique()
```

```python
# Function to extract CPU model (grouping all AMD CPUs under "AMD" and Intel Xeon under "Intel Other")
def extract_cpu_category(cpu):
    if 'Intel' in cpu:
        if 'Core i5' in cpu:
            return 'Intel Core i5'
        elif 'Core i7' in cpu:
            return 'Intel Core i7'
        elif 'Core i3' in cpu:
            return 'Intel Core i3'
        elif 'Core i9' in cpu:
            return 'Intel Core i9'
        elif 'Celeron' in cpu:
            return 'Intel Celeron'
        elif 'Xeon' in cpu:
            return 'Intel Other'  # Group Xeon under Intel Other
        else:
            return 'Intel Other'
    elif 'AMD' in cpu:
        return 'AMD'  # Group all AMD models under 'AMD'
    else:
        return 'Other'

# Apply the function to the 'Cpu' column and keep as object type
df['Cpu'] = df['Cpu'].apply(extract_cpu_category)

# Ensure the 'Cpu_Category' is of type object (which is default)
df['Cpu'] = df['Cpu'].astype('object')


```

```python
# Create 'TouchScreen' column: 'Yes' if the resolution contains 'Touchscreen', else 'No'
df['TouchScreen'] = df['ScreenResolution'].apply(lambda x: 'Yes' if 'Touchscreen' in str(x) else 'No')

# Create 'IPS' column: 'Yes' if the resolution contains 'IPS', else 'No'
df['IPS'] = df['ScreenResolution'].apply(lambda x: 'Yes' if 'IPS' in str(x) else 'No')
```
